```python
import requests
from bs4 import BeautifulSoup
import time
import csv

# Set headers to mimic a real browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
}

# Function to scrape reviews from a single page
def scrape_amazon_reviews(url):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")

    reviews = []

    # Extract review elements
    for review in soup.select("div[data-hook='review']"):
        try:
            # Extract review text
            review_text = review.select_one("span[data-hook='review-body']").text.strip()
            
            # Extract review rating
            rating = review.select_one("i[data-hook='review-star-rating']").text.strip().split(" ")[0]
            
            # Extract reviewer name
            reviewer_name = review.select_one("span.a-profile-name").text.strip()

            reviews.append({
                "reviewer_name": reviewer_name,
                "rating": rating,
                "review_text": review_text,
            })
        except AttributeError:
            # Skip if any element is missing
            continue

    return reviews

# Function to handle pagination
def scrape_all_reviews(product_id, max_pages=5):
    base_url = f"https://www.amazon.com/product-reviews/{product_id}"
    all_reviews = []

    for page in range(1, max_pages + 1):
        url = f"{base_url}/ref=cm_cr_arp_d_paging_btm_next_{page}?pageNumber={page}"
        print(f"Scraping page {page}...")
        reviews = scrape_amazon_reviews(url)
        if not reviews:
            print(f"No reviews found on page {page}. Stopping pagination.")
            break
        all_reviews.extend(reviews)

        # Be polite: add a delay between requests
        time.sleep(2)

    return all_reviews

# Function to save data to a CSV file
def save_to_csv(data, filename="amazon_reviews.csv"):
    if not data:
        print("No data to save. The reviews list is empty.")
        return

    keys = data[0].keys()  # Get the keys (column names) from the first review
    with open(filename, mode="w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=keys)
        writer.writeheader()  # Write the header row
        writer.writerows(data)  # Write all review rows
    print(f"Data saved to {filename}")

# Function to read and display data from a CSV file
def view_csv(filename="amazon_reviews.csv"):
    try:
        with open(filename, mode="r", encoding="utf-8") as file:
            reader = csv.DictReader(file)
            for row in reader:
                print(f"Reviewer: {row['reviewer_name']}")
                print(f"Rating: {row['rating']}")
                print(f"Review: {row['review_text']}")
                print("-" * 50)
    except FileNotFoundError:
        print(f"File {filename} not found. No data to display.")

# Example usage
if __name__ == "__main__":
    # Replace with the actual product ID (found in the product URL)
    product_id = "B08N5WRWNW"  # Example product ID for a pair of headphones
    reviews = scrape_all_reviews(product_id, max_pages=3)

    # Save the scraped data to a CSV file
    save_to_csv(reviews)

    # View the saved data
    view_csv()
```
