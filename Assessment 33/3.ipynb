```python
# Import Libraries
import pandas as pd
from geopy.geocoders import Nominatim
import folium
```

```python
# Load customer data from a CSV file
df = pd.read_csv('customer_locations.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Initialize the geocoder
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
# Function to get latitude and longitude from an address
def get_lat_lon(address):
    location = geolocator.geocode(address)
    return (location.latitude, location.longitude)
```

```python
# Display the updated DataFrame
print(df.head())
```

```python
# Create a map centered at the first customer's location
map_center = [df.iloc[0]['latitude'], df.iloc[0]['longitude']]
mymap = folium.Map(location=map_center, zoom_start=10)
```

```python
# Display the map
mymap
```

```python
# Add markers for each customer location
for index, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['name'],  # Display customer name in the popup
        icon=folium.Icon(color='blue')
    ).add_to(mymap)
```

```python
# Display the map with markers
mymap
```

```python
# Save the map to an HTML file
mymap.save('customer_locations_map.html')
```
